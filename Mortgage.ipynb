# Amortization Calculator
Enter values into the first cell (below). Do not modify the other cells.
Once you have entered in the values, simply run every cell by pressing Shift+Enter

```python
#Enter Principle Here (EX:500000)
L = 500000

#Enter Monthly Payment Here (EX:5000)
payment = 5000

#Enter Monthly Taxes, Fees and Insurance Cost Here (EX:500)
fee = 500

#Enter Annual Nominal Interest Rate As A Decimal Here(EX:0.05)
rate = 0.05
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
n=(-np.log(1-((L*rate/12)/(payment-fee))))/(np.log(1+rate/12))
Q=payment-fee
interest=rate/12
v=1/(1+(interest))
```

```python
def balance(time):
    power=np.power(v,(n-time))
    a=(1-power)/(rate/12)
    return(Q*a)

def principle(time):
    power=np.power(v,n+1-time)
    return(Q*power)

def interest(time):
    power=np.power(v,(n+1-time))
    return(Q*(1-power))

print("After{:4d} months you will owe ${:9.2f} with ${:7.2f} principle paid, ${:7.2f} interest paid and ${:6.2f} fees paid".format(0,balance(0),0,0,0))
for i in range (1,int(np.ceil(n))):
    print("After{:4d} months you will owe ${:9.2f} with ${:7.2f} principle paid, ${:7.2f} interest paid and ${:5.2f} fees paid".format(i,balance(i),principle(i),interest(i),fee))
    if(i%12==0):
        print("{:d}years passed".format(int(i/12)))
print("After{:4d} months your mortgage will be paid off".format(int(np.ceil(n))))
print("The total Interest Paid is ${:7.2f}".format(float(n*Q)-L))

x=list()
p=list()
for a in range(0,int(np.ceil(n))):
    x.append(a)
    p.append(balance(a))

plt.plot(x,p)
plt.xlabel("Number of Months")
plt.ylabel("Outstanding Principle Balance")
plt.show()
```
